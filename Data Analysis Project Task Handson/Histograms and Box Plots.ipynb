```python
import pandas as pd
import matplotlib.pyplot as plt
reviews = pd.read_csv("https://dsserver-prod-resources-1.s3.amazonaws.com/144/fandango_scores.csv?versionId=CLjyW02Qv.e_r2LCSRVOr4ahtnXfSlyC")
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
norm_reviews = reviews[cols]
value_counts = norm_reviews['Fandango_Ratingvalue'].value_counts()
fandango_distribution = value_counts.sort_index()
value_counts1 = norm_reviews['IMDB_norm'].value_counts()
imdb_distribution = value_counts.sort_index()
print(fandango_distribution)
```

```python
fig,ax = plt.subplots()
ax.hist(norm_reviews['Fandango_Ratingvalue'], range=(0, 5))
plt.show()
```

```python
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)
ax1.hist(norm_reviews['Fandango_Ratingvalue'], 20, range=(0, 5))
ax1.set_title('Distribution of Fandango Ratings')
ax2.hist(norm_reviews['RT_user_norm'], 20, range=(0, 5))
ax2.set_title('Distribution of Rotten Tomatoes Ratings')
ax3.hist(norm_reviews['Metacritic_user_nom'], 20, range=(0, 5))
ax3.set_title('Distribution of Metacritic Ratings')
ax4.hist(norm_reviews['IMDB_norm'], 20, range=(0, 5))
ax4.set_title('Distribution of IMDB Ratings')
ax1.set_ylim(0,50)
ax2.set_ylim(0,50)
ax3.set_ylim(0,50)
ax4.set_ylim(0,50)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.boxplot(norm_reviews['RT_user_norm'])
ax.set_ylim(0,5)
ax.set_xlabel('Rotten Tomatoes')
ax.set_xticklabels(norm_reviews['RT_user_norm'],rotation=90)
plt.show()
```

```python
fig,ax = plt.subplots()
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
ax.boxplot(norm_reviews[num_cols].values)
ax.set_ylim(0,5)
ax.set_xticklabels(num_cols,rotation=90)
plt.show()
```
