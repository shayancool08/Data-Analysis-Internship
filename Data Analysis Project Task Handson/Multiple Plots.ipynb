```python
import pandas as pd;
import matplotlib.pyplot as plt
```

```python
unrate = pd.read_csv("https://dsserver-prod-resources-1.s3.amazonaws.com/142/unrate.csv?versionId=pt.8VbQp6DWb4bjUyATUOXyQTCo5cMbs")
```

```python
unrate['DATE']=pd.to_datetime(unrate['DATE'])
```

```python
unrate.head(12)
```

```python
plt.plot()
```

```python
unrate_1=unrate.head(12)
```

```python
unrate_1.plot(x='DATE',y='VALUE')
plt.xticks(rotation=90)
plt.plot()
```

```python
unrate_1.plot(x='DATE',y='VALUE')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.xticks(rotation=90)
plt.plot()
```

```python
fig=plt.figure()
```

```python
plt.show()
```

```python
ax1.plot(unrate['DATE'][:12],unrate['VALUE'][:12])
```

```python
ax1.plot()
ax2.plot()
```

```python
fig = plt.figure(figsize=(12,12))
for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])
plt.show()
```

```python
fig = plt.figure(figsize=(6,3))
unrate['MONTH'] = unrate['DATE'].dt.month
plt.plot(unrate['MONTH'][:12],unrate['VALUE'][:12],c='red')
plt.plot(unrate['MONTH'][12:24],unrate['VALUE'][12:24],c='blue')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
unrate['MONTH'] = unrate['DATE'].dt.month
plt.plot(unrate['MONTH'][:12],unrate['VALUE'][:12],c='red')
plt.plot(unrate['MONTH'][12:24],unrate['VALUE'][12:24],c='blue')
plt.plot(unrate['MONTH'][24:36],unrate['VALUE'][24:36],c='green')
plt.plot(unrate['MONTH'][36:48],unrate['VALUE'][36:48],c='orange')
plt.plot(unrate['MONTH'][48:60],unrate['VALUE'][48:60],c='black')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
unrate['MONTH'] = unrate['DATE'].dt.month
plt.plot(unrate['MONTH'][:12],unrate['VALUE'][:12],c='red')
plt.plot(unrate['MONTH'][12:24],unrate['VALUE'][12:24],c='blue')
plt.plot(unrate['MONTH'][24:36],unrate['VALUE'][24:36],c='green')
plt.plot(unrate['MONTH'][36:48],unrate['VALUE'][36:48],c='orange')
plt.plot(unrate['MONTH'][48:60],unrate['VALUE'][48:60],c='black')
plt.legend(loc="upper left",labels=["1948","1949","1950","1951","1952"])
plt.show()
```
