```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv("cardioActivities.csv",parse_dates=True,index_col="Date")
df_run = df_activities[df_activities["Type"]=="Running"].copy()
df_run.drop(["Friend's Tagged","Notes"],inplace=True,axis=1)
df_walk = df_activities[df_activities["Type"]=="Walking"].copy()
df_walk.drop(["Friend's Tagged","Notes"],inplace=True,axis=1)
df_cycle = df_activities[df_activities["Type"]=="Cycling"].copy()
df_cycle.drop(["Friend's Tagged","Notes"],inplace=True,axis=1)
df_run.head()
```

```python
df_walk.head()
```

```python
df_cycle.head()
```

```python
df_run_walk_cycle = df_run.append([df_walk,df_cycle]).sort_index(ascending=False)
df_run_walk_cycle.head()
```

```python
dist_climb_cols = ['Distance (km)','Climb (m)']
df_totals = df_run_walk_cycle.groupby("Type")[dist_climb_cols].sum()
df_totals
```

```python
df_summary = df_run_walk_cycle.groupby("Type")['Average Speed (km/h)','Distance (km)','Climb (m)'].sum()
df_summary
```
