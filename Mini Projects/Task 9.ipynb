```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv("cardioActivities.csv",parse_dates=True,index_col="Date")
df_run = df_activities[df_activities["Type"]=="Running"].copy()
df_run.drop(["Friend's Tagged","Notes"],inplace=True,axis=1)
runs_subset_2015_2018 = df_run.loc['2018-12-31':'2015-01-01']
runs_subset_2015_2018
```

```python
df_run_hr_all = runs_subset_2015_2018['Average Heart Rate (bpm)']
df_run_hr_all
```

```python
fig, ax = plt.subplots(figsize=(8,5))
hr = [75,100,150,180]
zone_names = ["Normal","Fast","Very Fast","Abnormally fast"]
ax.hist(df_run_hr_all,bins=hr, color= "blue")
ax.xaxis.set(ticks=hr)
ax.set_xticklabels(labels=zone_names,rotation=-30,ha='left')
plt.show()
```
