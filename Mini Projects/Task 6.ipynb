```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv("cardioActivities.csv",parse_dates=True,index_col="Date")
df_run = df_activities[df_activities["Type"]=="Running"].copy()
df_run.drop(["Friend's Tagged","Notes"],inplace=True,axis=1)
runs_subset_2015_2018 = df_run.loc['2018-12-31':'2015-01-01']
runs_subset_2015_2018
```

```python
runs_distance = runs_subset_2015_2018['Distance (km)']
runs_hr = runs_subset_2015_2018['Average Heart Rate (bpm)']
runs_distance
```

```python
runs_hr
```

```python
fig,(ax1,ax2) = plt.subplots(2,sharex=True,figsize=(12,8))
runs_distance.plot(ax=ax1)
ax1.set(xlabel="Year",ylabel="Distance")
runs_hr.plot(ax=ax2)
ax2.set(xlabel="Year",ylabel="Heart rate")
ax2.axhline()
```
