```python
df_cycle = df_activities[df_activities["Type"]=="Cycling"].copy()
df_cycle
```

```python
df_cycle['Average Heart Rate (bpm)'].fillna(avg_hr_cycle['Average Heart Rate (bpm)'],inplace=True)
df_cycle
```
