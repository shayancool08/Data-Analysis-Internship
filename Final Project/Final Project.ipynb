```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Train.csv")
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
pd.to_numeric(df['Item_Weight'], errors ='ignore') 
df.head(10)
```

```python
df.dtypes
```

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(),inplace=True)
df.head(10)
```

```python
df['Outlet_Size'].fillna("No Size",inplace=True)
df.head(10)
```

```python
plt.figure(figsize=(15,10))
df.groupby('Item_Type')['Item_Outlet_Sales'].agg(['sum']).plot.bar()
plt.title("TOTAL SALES OF EACH ITEM TYPE")
plt.ylabel("Sum of sales")
plt.show()
```

```python
typebymaxw = df.groupby(['Item_Type']).agg(['sum']).reset_index()
typebymaxw.plot(x='Item_Type', y='Item_Weight',kind="barh",rot=0)
plt.title("ITEM TYPE BY TOTAL ITEM WEIGHT")
plt.xlabel("Sum of Item Weights")
plt.show()
```

```python
from matplotlib.gridspec import GridSpec
plt.figure(1, figsize=(20,10))
the_grid = GridSpec(2, 3)
plt.subplot(the_grid[0, 2], aspect=1, title='Outlet Location Type')
df['Outlet_Location_Type'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(the_grid[0, 1], aspect=1, title='Outlet Sizes')
df['Outlet_Size'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(the_grid[0, 0], aspect=1, title='Item fat Content')
df['Item_Fat_Content'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
sns.catplot(y="Outlet_Type", hue="Item_Type", kind="count",data=df,height=20);
plt.title("Different Outlet Type by different Item Types")
plt.show()
```

```python
typebymaxv = df.groupby(['Item_Type']).agg(['max']).reset_index()
typebymaxv.plot(x='Item_Type', y='Item_Visibility',kind="scatter",rot=90,color="red")
plt.title("ITEM TYPE BY MAX ITEM VISIBILITY")
plt.show()
```

```python
gross = df.groupby(['Outlet_Type']).agg(['sum']).reset_index()
gross.plot(x='Outlet_Type', y='Item_Outlet_Sales',kind="barh",rot=0)
plt.title("TOTAL SALES OF EACH OUTLET TYPE")
plt.xlabel("Sum of MRP")
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
reg = LinearRegression()
labels = df['Item_MRP']
train1 = df.drop(['Item_Identifier', 'Outlet_Identifier','Item_MRP','Item_Fat_Content','Item_Type','Outlet_Establishment_Year','Outlet_Size','Outlet_Type','Outlet_Location_Type'],axis=1)
train1['Item_Weight'].fillna(df['Item_Weight'].mean(),inplace=True)
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.1,random_state =2)
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
pred=reg.predict(x_test)
data1=pd.DataFrame({"Actual":y_test,"predicted":pred})
data1
```

```python
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test,pred)))
print(metrics.mean_absolute_error(y_test, pred))
print(metrics.r2_score(y_test, pred))
```
